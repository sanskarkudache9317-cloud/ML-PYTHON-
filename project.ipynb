```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
# X = input features
# y = target feature to predict

X = df.drop(['Eating_Disorders', 'Country', 'Code'], axis=1)
y = df['Eating_Disorders']
```

```python
scaler = StandardScaler()
X_model_norm = scaler.fit_transform(X)

# Convert target to array
y_model = y.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_model_norm, y_model, test_size=20, random_state=42
)

print("Shape of x_train : ", X_train.shape)
print("Shape of y_train : ", y_train.shape)
print("Shape of x_test : ", X_test.shape)
print("Shape of y_test : ", y_test.shape)
```

```python
Model = LinearRegression()
Model.fit(X_train, y_train)
```

```python
y_pred = Model.predict(X_test)
```

```python
print("Mean Absolute Error of Model is: ",
      metrics.mean_absolute_error(y_test, y_pred))

print("Mean Squared Error of Model is: ",
      metrics.mean_squared_error(y_test, y_pred))

print("Root Mean Squared of Model is: ",
      np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

print("R2 Score of Model is: ",
      metrics.r2_score(y_test, y_pred))
```

```python
k_fold = KFold(10)

scores = cross_val_score(
    Model,
    X_model_norm,
    y_model.ravel(),
    cv=k_fold,
    n_jobs=1
)

print(scores)
```

```python
x1 = df['Anxiety']
y1 = df['Depression']

# Scatter plot
plt.scatter(x1, y1, color='pink', label='Data')

# Smooth range
X_range = np.linspace(x1.min(), x1.max(), 100).reshape(-1, 1)

# Create a new PolynomialFeatures and LinearRegression model for this specific plot
degree = 2 # Assuming the same degree as before
poly_plot = PolynomialFeatures(degree=degree)
model_plot = LinearRegression()

# Fit the new polynomial features and model using only the 'Anxiety' data
X_anxiety_poly = poly_plot.fit_transform(x1.values.reshape(-1, 1))
model_plot.fit(X_anxiety_poly, y1)

# Transform the smooth range using the new polynomial features object
X_range_poly = poly_plot.transform(X_range)

# Predict using the new model
y_range_pred = model_plot.predict(X_range_poly)

# Plot curve
plt.plot(X_range, y_range_pred, color='blue', label='Polynomial Regression')

plt.xlabel('Anxiety')
plt.ylabel('Depression')
plt.title('Polynomial Regression Curve')
plt.legend()
plt.show()
```
